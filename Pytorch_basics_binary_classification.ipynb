```python
import torch.nn as nn
import pandas as pd
import numpy as np
import torch.optim as optim
import torch
```

```python
# generate the data

np.random.seed(42)   # same like random_state in ML ..... the X_data will be same everytime
X_data = np.random.uniform(low=1, high=100, size=100)
X_data
```

```python
# adding noise

noise= np.random.normal(loc=0, scale=0.5, size=100)  # loc--> whole data mean will be 0(mean=0) and scale(std deviation)=0.5
noise

# to make the data random we are adding noise
```

```python
y_data = ((X_data + noise) > 5).astype(int)
```

```python
y_data
```

```python
X_data
```

```python
# create a dataframe

data= {"feature":X_data, "target":y_data}
df= pd.DataFrame(data)
df
```

```python
# convert df to pytorch tensor

X = torch.tensor(df['feature'].values, dtype= torch.float32).view(-1,1)
Y = torch.tensor(df['target'].values, dtype= torch.float32).view(-1,1)
```

```python
X,Y
```

```python
# we need to write everything inside class while using pytorch

class LogisticRegression(torch.nn.Module):
    def __init__(self, input_size,hidden_size1, hidden_size2, output_size):
        super(LogisticRegression, self).__init__()

        self.layer1 = nn.Linear(input_size, hidden_size1)  # layer 1
        self.relu1= nn.ReLU()

        self.layer2 = torch.nn.Linear(hidden_size1, hidden_size2)  # layer 2
        self.sigmoid1= nn.Sigmoid()

        self.output = torch.nn.Linear(hidden_size2, output_size)  # op layer

    def forward(self, x):
        x= self.layer1(x)
        x = self.relu1(x)
        x = self.layer2(x)
        x = self.sigmoid1(x)
        x = self.output(x)
        return x
```

```python
input_size= 1  # number of featres
hidden_size1= 3
hidden_size2= 7
output_size= 1  # for binary classification- output size can be either 0 or 1

model= LogisticRegression(input_size, hidden_size1, hidden_size2, output_size)
```

```python
criterion= nn.BCEWithLogitsLoss()  # loss
optimizer = optim.SGD(model.parameters(), lr=0.01)   # optimizer
```

```python
num_epochs = 100
for epoch in range(num_epochs):
  # for forward popogation
  output= model(X)  # Y_predicted
  loss= criterion(output, Y)   # compare the output we calculated to Y(actual output)

  # for backward popogation
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  # print loss
  if (epoch+1) % 10 == 0:
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

  # print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')
```

```python
# to see the matrix parameters

with torch.no_grad():
  print(f'Trained Parameter: layer1 weight = {model.layer1.weight.data}, layer1 bias ={model.layer1.bias.data}')
  print(f'Trained Parameter: layer2 weight = {model.layer2.weight.data}, layer2 bias ={model.layer2.bias.data}')
```

```python
with torch.no_grad():
  prediction = torch.sigmoid(model(X)).numpy()
  print(prediction)
```

```python
df['prob'] = prediction
df['pred'] = (prediction > 0.5).astype(int)  # to convert the boolean values to int- True(0) False(1)
df['correct'] = (df['pred'] == df['target']).astype(int)

accuracy = np.mean(df['pred'] == df['target'])
accuracy
```

```python
# Pytorch With the drop-OUT ----
# we need to write everything inside class while using pytorch

class LogisticRegression(torch.nn.Module):
    def __init__(self, input_size,hidden_size1, hidden_size2, output_size):
        super(LogisticRegression, self).__init__()

        self.layer1 = nn.Linear(input_size, hidden_size1)  # layer 1
        self.relu1= nn.ReLU()
        self.dropout1 = nn.Dropout(0.5)

        self.layer2 = torch.nn.Linear(hidden_size1, hidden_size2)  # layer 2
        self.sigmoid1= nn.Sigmoid()
        self.dropout2 = nn.Dropout(0.5)

        self.output = torch.nn.Linear(hidden_size2, output_size)  # op layer

    def forward(self, x):
        x= self.layer1(x)
        x = self.relu1(x)
        x = self.dropout1(x)
        x = self.layer2(x)
        x = self.sigmoid1(x)
        x = self.dropout2(x)
        x = self.output(x)
        return x
```

```python
input_size= 1  # number of features
hidden_size1= 3
hidden_size2= 7
output_size= 1  # for binary classification- output size can be either 0 or 1

model= LogisticRegression(input_size, hidden_size1, hidden_size2, output_size)
```

```python
criterion= nn.BCEWithLogitsLoss()  # loss
optimizer = optim.SGD(model.parameters(), lr=0.01)   # optimizer
```

```python
num_epochs = 100
for epoch in range(num_epochs):
  # for forward popogation
  output= model(X)  # Y_predicted
  loss= criterion(output, Y)   # compare the output we calculated to Y(actual output)

  # for backward popogation
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  # print loss
  if (epoch+1) % 10 == 0:
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')
```
