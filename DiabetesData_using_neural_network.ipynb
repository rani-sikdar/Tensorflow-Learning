```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
from sklearn.model_selection import train_test_split
```

```python
# dataset link= https://www.kaggle.com/code/melikedilekci/diabetes-dataset-for-beginners/input

data= pd.read_csv('/content/diabetes.csv')
data.head()
```

```python
X= data.iloc[:, 0:8]
y= data['Outcome']

print(X.shape)
print(y.shape)
```

```python
X= data.iloc[:, 0:8]
y= data.iloc[:,8]

print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
X_test
```

```python
# build architecture now

model= tf.keras.Sequential()

# create and add the layers
# hidden layer 1
model.add(tf.keras.layers.Dense(10, activation='relu'))

# hidden layer 2
model.add(tf.keras.layers.Dense(20, activation= 'sigmoid'))

# output layer
model.add(tf.keras.layers.Dense(1))   # on the same node we will get either 0 or 1 value for each records thats why dense(1)

# dense will have 2 neuraons as this is a binary problem. Outcome- 0/1
```

```python
optimizer= tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=50)
```

Increasing the accuracy

we have to limit the output otherwise the output could be +infinity to -infinity

so we have to find probablity on loss function

```python
# build architecture now

model= tf.keras.Sequential()

# create and add the layers

# hidden layer 1
model.add(tf.keras.layers.Dense(10, activation='relu'))

# hidden layer 2
model.add(tf.keras.layers.Dense(20, activation= 'sigmoid'))

# output layer
model.add(tf.keras.layers.Dense(1, activation= 'sigmoid'))  # using sigmoid to inc probablity and dense is 1
```

```python
optimizer= tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=50, validation_data=(X_test, y_test))
```
