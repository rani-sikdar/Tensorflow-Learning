```python
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
tf.__version__
```

```python
dataset = {
    'feature 1': np.random.rand(100),
    'feature 2': np.random.rand(100),
    'target': 3 * np.random.rand(100) + 2 * np.random.rand(100) + 1 + 0.1 * np.random.rand(100)
}
```

```python
dataset
```

```python
df = pd.DataFrame(dataset)
df.head()
```

```python
x_train = df[['feature 1', 'feature 2']].values
y_train = df.target.values
```

```python
model = tf.keras.Sequential()

model.add(tf.keras.layers.Dense(20, activation= 'relu'))
model.add(tf.keras.layers.Dense(30, activation= 'sigmoid'))
model.add(tf.keras.layers.Dense(1, activation= 'sigmoid'))
```

```python
model.compile(optimizer= 'sgd', loss='mse',metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs= 50)
```

```python
model.compile(optimizer= 'adam', loss= 'mse')

model.fit(x_train, y_train, epochs= 50, batch_size = 32)
```

```python
# optimizer

op = tf.keras.optimizers.Adam(learning_rate= 0.001)
```

```python
model.compile(optimizer= op, loss='mse')
model.fit(x_train, y_train, epochs=50, batch_size=32)
```

```python
df = pd.read_csv('/content/diabetes.csv')
df.head()
```

```python
x = df.iloc[:, 0:8]
y = df['Outcome']
print(x.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 101)
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(10, activation='relu'))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(20, activation='sigmoid'))

model.add(tf.keras.layers.Dense(1, activation='sigmoid')) #op
```

```python
op= tf.keras.optimizers.Adam(learning_rate =0.001)
model.compile(optimizer= op, loss= 'mse', metrics= ['accuracy'])
```

```python
EPOCHS = 50
model.fit(x_train, y_train, epochs= EPOCHS, batch_size=20)
```

```python
op= tf.keras.optimizers.Adam(learning_rate =0.001)
model.compile(optimizer=op, loss= 'binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs= EPOCHS, batch_size=30,validation_data=(x_test, y_test))
```

```python
op= tf.optimizers.SGD(learning_rate= 0.001)
model.compile(optimizer=op, loss='binary_crossentropy', metrics= ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs= EPOCHS, batch_size= 20,validation_data= (x_test, y_test))
```
